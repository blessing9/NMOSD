### 相关性分析

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt
import seaborn as sns
from minepy import MINE
import xgboost
import eli5
from eli5.sklearn import PermutationImportance


def corr_columns(column1, column2, method):
    """
    计算两列数据的相关性（去掉缺失值-1）

    参数
    -------
    column1&column2:list
        对应两组变量
    method:str
        相关系数类型

    返回
    -------
    a:int
        参与计算相关系数的行数
    b:float
        相关系数
    """
    temp1,temp2 = [],[]
    for i in range(len(column1)):
        if column1[i] != -1 and column2[i] != -1:
            temp1.append(column1[i])
            temp2.append(column2[i])
    temp = pd.DataFrame({'A':temp1,'B':temp2})
    return len(temp1),temp.corr(method=method).iloc[0][1]
def corr(df, method):
    """
    计算pandas dataframe df中列与列的相关性（df中-1表示缺失值，不参与计算）

    参数
    -------
    df:pandas.DataFrame
        所有数据
    method:str
        相关系数类型

    返回
    -------
    result:pandas.DataFrame
        相关性计算结果
    """
    title = list(df.columns)
    result = [[] for i in title]
    for i in range(len(title)):
        for j in range(len(title)):
            if method == 'mic':
                a,b = mic(df[title[i]],df[title[j]])
            else:
                a,b = corr_columns(list(df[title[i]]), list(df[title[j]]), method)
            result[i].append(b)
    result = pd.DataFrame(data=result,index=title,columns=title)
    return result

def X2(v1,v2):
    """
    对两个类别变量进行卡方检验,画出列联表及柱状图

    参数
    -------
    v1:pandas.Series
        第一个变量的所有值,-1代表缺失
    v2:pandas.Series
        第二个变量的所有值,-1代表缺失
    
    返回
    ------
    result:tuple
        (卡方值,p值,自由度)
    """
    temp1, temp2 = [],[]
    for i in range(len(v1)):
        if v1[i] != -1 and v2[i] != -1:
            temp1.append(v1[i])
            temp2.append(v2[i])
    
    #计算列联表
    v1 = pd.Series(data=temp1,name=v1.name)
    v2 = pd.Series(data=temp2,name=v2.name)
    c_table = pd.crosstab(v1,v2,margins=True)
    print('列联表:')
    print(c_table)
    #卡方检验
    obs = np.array(c_table.iloc[:-1,:-1])
    x2 = chi2_contingency(obs)
    print('卡方值:%.2f,p值:%.2f,自由度:%i,expected_freq:%s'%x2)

def mic(column1, column2):
    """
    计算两个变量的最大互信息值

    参数
    ------
    column1:pandas.Series
        第一个变量，-1表示缺失值
    column2:pandas.Series
        第二个变量，-1表示缺失值
    
    返回
    -----
    result:float
        两个变量的最大互信息值
    """
    temp1,temp2 = [],[]
    for i in range(len(column1)):
        if column1[i] != -1 and column2[i] != -1:
            temp1.append(column1[i])
            temp2.append(column2[i])
    m = MINE()
    m.compute_score(temp1,temp2)
    return len(temp1), m.mic()

def xgboost_feature(x, y):
    """
    使用XGboost计算feature importance,排序并输出

    参数
    -----
    x:pandas.DataFrame
        决策变量
    y:pandas.Series
        预测变量

    """
    model = xgboost.XGBClassifier()
    clf = model.fit(x, y)
    a = clf.feature_importances_
    features = pd.DataFrame(sorted(zip(a, x.columns), reverse=True))
    features.to_csv('temp.csv')
    print(features)
    return features
    perm = PermutationImportance(model, random_state = 1).fit(x,y)
    eli5.show_weights(perm)
```

#### 序列相关性分析

```python

names = ['id', 'time', 'gender'] + [f'medicine-{i}' for i in range(6)]+['interval']

df = pd.read_csv('data/value_points_corr.csv')
df = df.drop(names, axis=1)
columns = df.columns
columns
```

```python
seq_index = open('data/point_sequences.csv','r').readlines()
data = df.iloc
names = columns[:-1]
result = []
for m in ['pearson', 'spearman','kendall','mic']:
    temp1,temp2 = [],[]
    for n in names:
        items,events = [],[]
        for line in seq_index:
            line = line.strip().split(',')
            id = int(line.pop(0))
            event = eval(line.pop(-1))
            line = [int(i) for i in line]

            if event == -1:
                continue

            temp = []
            for i in line:
                if data[i][n] != -1:
                    temp.append(data[i][n])
            if len(temp)>4:
                items.append(np.mean(temp))
                events.append(event)
        if m == 'mic':
            a, b = mic(items,events)
        else:
            a,b = corr_columns(items,events,m)
        print(f'{n}:\t{a}\t{b}')
        if a > 1000:
            temp1.append(n)
            temp2.append(b)
    s = pd.Series(data=temp2, index=temp1, name=m)
    result.append(s)
```

#### 静态变量（性别，第一次发作信息，出生月份）

```python
from data_make import patient
from utils import dis
import pandas as pd

patients_f = open('data/washed_sheet/patients.csv')
patients = []
for line in patients_f:
    patients.append(patient(line.strip()))
```

```python
#首发后一年内复发率
t,b = 0,0
for p in patients:
    span = dis(p.first_breakout[0], p.last_record())
    if len(p.breakout) == 1:
        t += int(span>=365)
    else:
        t += 1
        b += int(dis(p.breakout[0][0],p.breakout[1][0])<365)
print(f'{float(b)/t}:{b}/{t}')
```

```python
### 计算病人的静态变量（性别、年龄、ocb, 首次发作年龄，首次发作临床信息与首次发作后一年内是否复发、平均发作间隔、发作次数（平均）的相关性）
title = ['gender','birth_month','age', 'ocb']+ [f'symptom_{i}' for i in range(1,5)] + ['outcome', 'breakout_num','breakout_interval','event']
data = []
for p in patients:
    #平均发作次数
    span = float(dis(p.breakout[0][0],p.last_record()))/365
    if span < 3:   #只计算周期大于3年的病人静态变量与平均复发次数的相关性
        breakout_num = -1
    else:
        breakout_num = len(p.breakout)/span
    #平均发作时间间隔
    breakout_times = [b[0] for b in p.breakout]
    breakout_interval = [dis(breakout_times[i],breakout_times[i+1]) for i in range(len(breakout_times) - 1)]
    breakout_interval = sum(breakout_interval)/len(breakout_interval) if len(breakout_interval) != 0 else -1

    #首次发作后一年内复发情况
    if len(p.breakout) == 1:
        event = 0 if span >= 1 else -1
    else:
        event = 1 if dis(p.breakout[0][0],p.breakout[1][0])<365 else 0
    row = [int(p.gender), int(p.birthday[5:7]), dis(p.birthday,p.first_breakout[0])//365, p.ocb] + [int(i) for i in p.first_breakout[1:5]] + [float(p.first_breakout[9])] + [breakout_num,breakout_interval,event]
    data.append(row)
data = pd.DataFrame(data=data,columns=title)

method = 'mic'
result = corr(data,method)
result.to_csv(f'result/correlation_analysis/corr_static_{method}.csv')
```
